```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("job_descriptions_cleaned.csv")

# Merge the dataset with headers
merged_data = []
for index, row in df.iterrows():
    merged_data.append(f"{row['Job Title']}: The Experience is {row['Experience']}, Qualifications is {row['Qualifications']}, and Job Description is {row['Job Description']}.")

# Save the merged data to a new txt file
with open("merged_data.txt", "w") as file:
    for data in merged_data:
        file.write(data + "\n")
```

```python
df = pd.read_csv('academic_routes_final.csv')
```

```python
df.head()
```

```python
# replace true false with required and not required for columns Bachelors, Masters, and PhD
df['Bachelors'] = df['Bachelors'].replace({True: 'offered', False: 'not offered'})
df['Masters'] = df['Masters'].replace({True: 'offered', False: 'not offered'})
df['PhD'] = df['PhD'].replace({True: 'offered', False: 'not offered'})
```

```python
# Merge the dataset with headers
merged_data = []
for index, row in df.iterrows():
    merged_data.append(f"{row['Program']}: {row['Description']} Subject areas are {row['Subject Areas']}. Bachelor's degree is {row['Bachelors']}. Master's degree is {row['Masters']}. Doctoral degree is {row['PhD']}.")

# Save the merged data to a new txt file
with open("merged_data_academic.txt", "w") as file:
    for data in merged_data:
        file.write(data + "\n")
```
